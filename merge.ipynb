```python
import numpy as np
import pandas as pd
import os
from featuretools.selection import remove_single_value_features
```

```python
datasets = os.listdir('Data')
```

```python
train_csvs = [data for data in datasets if "train" in data]
test_csvs = [data for data in datasets if "test" in data]
```

```python
final_df = pd.DataFrame()
for df in train_csvs:
    data = pd.read_csv("Data/{}".format(df), index_col = 0)
    final_df = pd.concat([final_df, data], axis = 1, )
```

```python
final_df['country']
# check we have merged correctly 
for i in range(len(final_df['country'])):
    row = list(np.array(final_df['country'].iloc[i]))
    if len(set(row)) > 1:
        print("FALSE")

for i in range(len(final_df['year'])):
    row = list(np.array(final_df['year'].iloc[i]))
    if len(set(row)) > 1:
        print("FALSE")
```

```python
metrics = [train.split('_')[0] for train in train_csvs]
final_df[metrics]
```

```python
metrics
```

```python
env = ['fwateruse',
 'airpollution',
 'wasterwater',
 'fishstock',
 'marineprotect',
 'improved_drinkwater',
 'pestreg',
 'terrestprotect']
gov = ['elfcr',
 'trang',
 'womenparl',
 'vapvt',
 'polity2',
 'polrights',
 'geff',
 'rqual',
 'ticpi',
 'judi',
 'rlaw']
```

```python
final_df = final_df.loc[:,~final_df.columns.duplicated()]
```

```python
final_df.head()
```

```python
for i in final_df.columns:
    if i.find('_year') > -1:
        final_df = final_df.drop([i], axis = 1)
```

```python
final_df = remove_single_value_features(final_df)
```

```python
final_df['prosperity_score'] = 0
for metric in metrics:
    final_df['prosperity_score'] += final_df[metric]
```

```python
final_df['prosperity_score'] = final_df['prosperity_score'] / len(metrics)
```

```python
final_df.to_csv('merged.csv')
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/Andrewl7127/UCSD-DataHacks-2021/main/Data/merged.csv')
```

```python
columns = [i for i in df.columns if 'rank' in i]
```

```python
['country', 'year', 'prosperity_score', 'isocode'] + columns 
```

```python
subset = df[['country','year','prosperity_score','isocode','rank_educ','rank_soci','rank_heal','rank_pers','rank_busi','rank_econ','rank_safe','rank_gove','rank_envi']]
```

```python
v1 = subset.groupby(['country','year','isocode','rank_educ','rank_soci','rank_heal','rank_pers','rank_busi','rank_econ','rank_safe','rank_gove','rank_envi']).agg({'prosperity_score':sum}).reset_index()
```

```python
v1['index'] = v1.index
```

```python
v1.columns
```

```python
def get_percent_change(val, percentized):
    nation = v1.loc[v1['index'] == val, 'country'].values[0]
    important_index = v1[(v1['country'] == nation) & (v1['year'] == 2007)]['index'].values[0]
    
    val1 = v1.loc[v1['index'] == important_index, percentized].values[0]
    val2 = v1.loc[v1['index'] == (val), percentized].values[0]
    return ((val2 - val1) / val1) * 100
```

```python
v1['%_change_prosperity'] = v1['index'].apply(get_percent_change, percentized = 'prosperity_score')
v1['%_change_reduc'] = v1['index'].apply(get_percent_change, percentized = 'rank_educ')
v1['%_change_rsoci'] = v1['index'].apply(get_percent_change, percentized = 'rank_soci')
v1['%_change_rheal'] = v1['index'].apply(get_percent_change, percentized = 'rank_heal')
v1['%_change_rpers'] = v1['index'].apply(get_percent_change, percentized = 'rank_pers')
v1['%_change_rbusi'] = v1['index'].apply(get_percent_change, percentized = 'rank_busi')
v1['%_change_recon'] = v1['index'].apply(get_percent_change, percentized = 'rank_econ')
v1['%_change_rsafe'] = v1['index'].apply(get_percent_change, percentized = 'rank_safe')
v1['%_change_rgove'] = v1['index'].apply(get_percent_change, percentized = 'rank_gove')
v1['%_change_renvi'] = v1['index'].apply(get_percent_change, percentized = 'rank_envi')
```
